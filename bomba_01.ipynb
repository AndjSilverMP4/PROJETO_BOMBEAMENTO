```python
# Criando função para ciclar um motor

import time
from datetime import datetime, timedelta

class MotorController:
    def __init__(self):
        self.motores = {
            '01': True,  # True indica que o motor está funcionando
            '02': True
        }
        self.current_motor = '01'
        self.manual_mode = False

    def check_motor_status(self):
        return self.motores[self.current_motor]

    def switch_motor(self):
        if self.current_motor == '01':
            self.current_motor = '02'
        else:
            self.current_motor = '01'

    def run_manual(self):
        while True:
            command = input("Digite 'on' para ligar, 'off' para desligar, 'exit' para sair: ").strip().lower()
            if command == 'on':
                self.motores[self.current_motor] = True
                print(f"Motor {self.current_motor} ligado manualmente.")
            elif command == 'off':
                self.motores[self.current_motor] = False
                print(f"Motor {self.current_motor} desligado manualmente.")
            elif command == 'exit':
                print("Saindo do modo manual.")
                break
            else:
                print("Comando inválido.")

    def run_automatic(self):
        end_time = datetime.now() + timedelta(hours=6)
        while datetime.now() < end_time:
            if not self.check_motor_status():
                print(f"Motor {self.current_motor} falhou. Trocando para o motor.")
                self.switch_motor()
                if not self.check_motor_status():
                    print(f"Motor {self.current_motor} também falhou. Tentando o outro motor novamente.")
                    self.switch_motor()
            print(f"Motor {self.current_motor} está operando.")
            time.sleep(10)  # Ajustar para tempo de verificação desejado

def main():
    controller = MotorController()

    mode = input("Escolha o modo ('manual' ou 'automatic'): ").strip().lower()
    if mode == 'manual':
        controller.run_manual()
    elif mode == 'automatic':
        controller.run_automatic()
    else:
        print("Modo inválido.")

if __name__ == "__main__":
    main()
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np

class Motor:
    def __init__(self, power, voltage, current, fs, tp, efficiency, ambient_temp):
        self.power = power
        self.voltage = voltage
        self.current = current
        self.fs = fs
        self.tp = tp
        self.efficiency = efficiency
        self.ambient_temp = ambient_temp
        self.frequency = 0

    def start_ramp(self, target_freq, ramp_time):
        steps = int(ramp_time / 40)
        freq_step = target_freq / steps
        for i in range(steps):
            self.frequency += freq_step
            time.sleep(40)
        self.frequency = target_freq

motor = Motor(3, [220, 380, 440], [15.0, 8.68, 7.50], 1.15, 0.89, 88.5, 40)

class Bomb:
    def __init__(self, name):
        self.name = name
        self.running = False
        self.failed = False
        self.frequency = 0

    def start(self):
        self.running = True
        motor.start_ramp(35, 40)  # Atingir 35 hz em 40 segundos
        self.frequency = motor.frequency

    def stop(self):
        self.running = False

bomb1 = Bomb("Bomb 01")
bomb2 = Bomb("Bomb 02")

def cycle_pumps():
    global bomb1, bomb2
    cycle_time = 180  # 3 minutos em segundos
    while True:
        if not bomb1.failed:
            bomb1.start()
            print(f"{bomb1.name} está em operação a {bomb1.frequency} Hz.")
            time.sleep(cycle_time)
            bomb1.stop()
            print(f"{bomb1.name} parou.")
        elif not bomb2.failed:
            bomb2.start()
            print(f"{bomb2.name} está em operação a {bomb2.frequency} Hz.")
            time.sleep(cycle_time)
            bomb2.stop()
            print(f"{bomb2.name} parou.")
        else:
            print("Ambas as bombas falharam.")
            break

# Iniciar o ciclo de bombas
cycle_pumps()

def manual_frequency_control():
    global bomb1, bomb2
    while True:
        freq = float(input("Defina a frequência manualmente (0-60 Hz): "))
        if 0 <= freq <= 60:
            motor.frequency = freq
            print(f"Frequência ajustada para {freq} Hz.")
        else:
            print("Frequência inválida.")

# Controle manual de frequência
manual_frequency_control()

def plot_motor_parameters():
    frequencies = np.arange(0, 61, 1)
    currents = np.interp(frequencies, [0, 60], [motor.current[0], motor.current[2]])
    powers = np.interp(frequencies, [0, 60], [0, motor.power])

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(frequencies, currents, label="Corrente (A)")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Corrente (A)")
    plt.title("Corrente vs. Frequência")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(frequencies, powers, label="Potência (kW)")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Potência (kW)")
    plt.title("Potência vs. Frequência")
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plotar parâmetros do motor
plot_motor_parameters()
```

```python
import time
import tkinter as tk
from tkinter import ttk
import matplotlib.pyplot as plt
import numpy as np

class Motor:
    def __init__(self, power, voltage, current, fs, tp, efficiency, ambient_temp):
        self.power = power
        self.voltage = voltage
        self.current = current
        self.fs = fs
        self.tp = tp
        self.efficiency = efficiency
        self.ambient_temp = ambient_temp
        self.frequency = 0

    def start_ramp(self, target_freq, ramp_time):
        steps = int(ramp_time / 40)
        freq_step = target_freq / steps
        for i in range(steps):
            self.frequency += freq_step
            time.sleep(40)
        self.frequency = target_freq

motor = Motor(3, [220, 380, 440], [15.0, 8.68, 7.50], 1.15, 0.89, 88.5, 40)

class Bomb:
    def __init__(self, name):
        self.name = name
        self.running = False
        self.failed = False
        self.frequency = 0

    def start(self):
        self.running = True
        motor.start_ramp(35, 40)
        self.frequency = motor.frequency

    def stop(self):
        self.running = False

bomb1 = Bomb("Bomb 01")
bomb2 = Bomb("Bomb 02")

def cycle_pumps():
    global bomb1, bomb2
    cycle_time = 180  # 3 minutos em segundos
    while True:
        if not bomb1.failed:
            bomb1.start()
            print(f"{bomb1.name} está em operação a {bomb1.frequency} Hz.")
            time.sleep(cycle_time)
            bomb1.stop()
            print(f"{bomb1.name} parou.")
        elif not bomb2.failed:
            bomb2.start()
            print(f"{bomb2.name} está em operação a {bomb2.frequency} Hz.")
            time.sleep(cycle_time)
            bomb2.stop()
            print(f"{bomb2.name} parou.")
        else:
            print("Ambas as bombas falharam.")
            break

def manual_frequency_control():
    global bomb1, bomb2
    while True:
        freq = float(input("Defina a frequência manualmente (0-60 Hz): "))
        if 0 <= freq <= 60:
            motor.frequency = freq
            print(f"Frequência ajustada para {freq} Hz.")
        else:
            print("Frequência inválida.")

def plot_motor_parameters():
    frequencies = np.arange(0, 61, 1)
    currents = np.interp(frequencies, [0, 60], [motor.current[0], motor.current[2]])
    powers = np.interp(frequencies, [0, 60], [0, motor.power])

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(frequencies, currents, label="Corrente (A)")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Corrente (A)")
    plt.title("Corrente vs. Frequência")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(frequencies, powers, label="Potência (kW)")
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Potência (kW)")
    plt.title("Potência vs. Frequência")
    plt.legend()

    plt.tight_layout()
    plt.show()

class CLPApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Controlador Lógico Programável")

        self.frame = ttk.Frame(root, padding="10")
        self.frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        self.bomb1_button = ttk.Button(self.frame, text="Iniciar Bomb 01", command=self.start_bomb1)
        self.bomb1_button.grid(row=0, column=0, padx=5, pady=5)

        self.bomb2_button = ttk.Button(self.frame, text="Iniciar Bomb 02", command=self.start_bomb2)
        self.bomb2_button.grid(row=0, column=1, padx=5, pady=5)

        self.fail_bomb1_button = ttk.Button(self.frame, text="Falha Bomb 01", command=self.fail_bomb1)
        self.fail_bomb1_button.grid(row=1, column=0, padx=5, pady=5)

        self.fail_bomb2_button = ttk.Button(self.frame, text="Falha Bomb 02", command=self.fail_bomb2)
        self.fail_bomb2_button.grid(row=1, column=1, padx=5, pady=5)

        self.freq_label = ttk.Label(self.frame, text="Frequência (Hz):")
        self.freq_label.grid(row=2, column=0, padx=5, pady=5)

        self.freq_entry = ttk.Entry(self.frame)
        self.freq_entry.grid(row=2, column=1, padx=5, pady=5)

        self.set_freq_button = ttk.Button(self.frame, text="Definir Frequência", command=self.set_frequency)
        self.set_freq_button.grid(row=3, column=0, columnspan=2, padx=5, pady=5)

        self.plot_button = ttk.Button(self.frame, text="Plotar Parâmetros do Motor", command=plot_motor_parameters)
        self.plot_button.grid(row=4, column=0, columnspan=2, padx=5, pady=5)

    def start_bomb1(self):
        bomb1.start()
        print(f"{bomb1.name} está em operação.")

    def start_bomb2(self):
        bomb2.start()
        print(f"{bomb2.name} está em operação.")

    def fail_bomb1(self):
        bomb1.failed = True
        print(f"{bomb1.name} falhou.")

    def fail_bomb2(self):
        bomb2.failed = True
        print(f"{bomb2.name} falhou.")

    def set_frequency(self):
        freq = float(self.freq_entry.get())
        if 0 <= freq <= 60:
            motor.frequency = freq
            print(f"Frequência ajustada para {freq} Hz.")
        else:
            print("Frequência inválida.")

root = tk.Tk()
app = CLPApp(root)
root.mainloop()
```
